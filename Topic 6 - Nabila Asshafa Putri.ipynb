# Join

Using two of the dataframes below, answer the questions :

```python
df1 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                   'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})
                   
df2=pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                      'B': ['B0', 'B1', 'B2']})
```

```python
#import libraries here
import pandas as pd
```

### Create a dataframe (df3) which consists both dataframe inner-joined by column "Key"

```python
#code goes here
df3 = pd.merge(df1,df2,on='key')
df3
```

### Create a dataframe (df4) which consists of the result of df1 and df2 left-joined by column "Key"

```python
#code goes here
df4 = pd.merge(df1,df2,on='key', how='left')
df4
```

#### 1. Replace key "K2" into "K3" on df1
#### 2. Left join it with df2 

```python
#code goes here
df1['key']=df1['key'].str.replace('K2','K3')
df1

df5 = pd.merge(df1,df2,on='key', how='left')
df5
```

# Dataframe Processing


Create a dictionary and convert it to a dataframe called "Customer_df"


name=['Anna','Jason','Cindy']

age=['22','23','21']

gender=['F','M','F']




```python
#code goes here

Customer_df = {
  "nama" : ['Anna','Jason','Cindy'],
  "age" : [22, 23, 21],
  "gender" : ['F','M','F']
}

newdf = pd.DataFrame(Customer_df)
newdf
```

### Delete the "age" column

```python
#code goes here
newdf.drop('age', axis=1)
```

### Rename the "name" column into "first name"

```python
#code goes here
renamed_columns={'nama':'first name'}

newdf = newdf.rename(columns=renamed_columns)
newdf
```

### Sort the dataframe by age, descending

```python
#code goes here

newdf.sort_values('age', ascending=False)
```
