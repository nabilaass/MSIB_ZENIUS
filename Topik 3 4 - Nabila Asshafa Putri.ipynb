# Basic Data Operation

You sold 20 pieces of electronics at the price of 12000. The customer gets a total discount of 10%.

Below are those variables:

1.   price=12000
2.   quantity_sold=20
3.   discount=10%


Using Python code, write the code the find the total money that the customer to pay. (Points : 15)

```python
price=12000
quantity_sold=20
discount= 10/100

buy = quantity_sold*price
print ('the customer buy=',buy)

disc = buy * discount
print ('the customer gets a discount= ',disc)

pay = buy - disc
print ('total money that the customer to pay=', pay) 

```
