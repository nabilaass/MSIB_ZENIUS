### Task 3: What is the the distribution of Age, Sex, BMI and Y variables?

```python
# Coinflips example - What is the population?

coinflips = np.random.binomial(n=1, p=.5, size=10)

df = pd.DataFrame({'AGE': coinflips})

df['AGE'].plot.density()

print(df.shape)
df.head()
```

Variabel "age" menunjukkan kurva berdistribusi normal karena panjang sebelah kanan dan kiri sama. Pada grafik distribusi normal, mean (rata-rata), median, dan modus semuanya sama

```python
# Coinflips example - What is the population?

coinflips = np.random.binomial(n=1, p=.5, size=10)

df = pd.DataFrame({'SEX': coinflips})

df['SEX'].plot.density()

print(df.shape)
df.head()
```

Variabel "sex" menunjukkan grafik miring ke kanan yang artinya berdistribusi right skew/positive skew.
Nilai modus adalah titik tertinggi dari plot ini, sedangkan median dan mean jatuh di sebelah kanannya (atau, secara visual, di sebelah kanan puncak).
